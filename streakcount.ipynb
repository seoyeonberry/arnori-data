```python
data = {
    'userId': [
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        'aea35790-0af2-11e9-a714-853bad31efc0', 
        
        'bea35790-0af2-11e9-a714-853bad31efc0', 
        'bea35790-0af2-11e9-a714-853bad31efc0', 
        'bea35790-0af2-11e9-a714-853bad31efc0', 
        
        'cea35790-0af2-11e9-a714-853bad31efc0', 
        'cea35790-0af2-11e9-a714-853bad31efc0', 
        'cea35790-0af2-11e9-a714-853bad31efc0', 
        'cea35790-0af2-11e9-a714-853bad31efc0', 
    ],
    't': [
        pd.Timestamp('2018-04-23'), 
        pd.Timestamp('2018-04-24'), 
        pd.Timestamp('2018-05-02'), 
        pd.Timestamp('2018-12-31'), 
        pd.Timestamp('2019-01-01'), 
        pd.Timestamp('2019-01-02'), 
        
        pd.Timestamp('2018-05-01'), 
        pd.Timestamp('2018-05-01'), 
        pd.Timestamp('2018-05-05'),
        
        pd.Timestamp('2018-05-01'), 
        pd.Timestamp('2018-05-02'), 
        pd.Timestamp('2018-05-03'), 
        pd.Timestamp('2018-05-06')
    ]
}
df = (pd.DataFrame(data)
        .groupby(by=['userId', 't'])
        .size()
        .reset_index(name='totalCheckIns'))
```

```python
df
```

```python
gids = (df[['userId', 't']]
        .groupby('userId')
        .diff()
        .t
        .dt
        .days
        .ne(1)
        .cumsum()
       )

df['streakGroupId'] = gids
```

```python
df
```

```python
df['streak'] = df.groupby('streakGroupId').cumcount() + 1
```

```python
df
```

```python
gb = df.groupby(['userId'])
counts = gb.totalCheckIns.sum().to_frame()

stats = (counts
    .join(gb.agg({'streak': 'max'}).rename(columns={'streak': 'bestStreak'}))
    .join(gb.agg({'streak': 'last'}).rename(columns={'streak': 'lastStreak'}))
    .join(gb.agg({'t': 'last'}).rename(columns={'t': 'lastCheckedInAt'}))
    .reset_index())

stats
```

```python
stats['currentStreak'] = stats['lastStreak']
stats.loc[pd.datetime.today() - stats['lastCheckedInAt'] > pd.offsets.Day(2), 'currentStreak'] = 0
stats
```
